```python
import numpy as np
import matplotlib.pyplot as plt

# Load the CSV file into a NumPy array
data = np.loadtxt('outputspace.csv', delimiter=',')

# Get the unique values in column 1
unique_values = np.unique(data[:, 0])

# Dictionary to store the closest values for each unique value in column 1
closest_values = {}

for value in unique_values:
    # Filter rows where column 1 matches the current unique value
    filtered_data = data[data[:, 0] == value]

   # Extract the second column of the filtered data
    second_column = filtered_data[:, 1]

    # Extract the last column of the filtered data
    last_column = filtered_data[:, -1]
    
    # Find the closest value to 0.1198 in the second column
    closest_value = second_column[np.abs(last_column - 0.1198).argmin()]
    
    # Store the closest value in the dictionary
    if 0.11 < last_column[np.abs(last_column - 0.1198).argmin()] and last_column[np.abs(last_column - 0.1198).argmin()] < 0.13:
        closest_values[value] = closest_value

    #closest_values[value] = closest_value

# Convert the dictionary into a list of tuples
closest_values_list = [(key, value) for key, value in closest_values.items()]

# Convert the list of tuples into a NumPy array
line = np.array(closest_values_list)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(line[:,0],line[:,1])
plt.xscale('log')
plt.yscale('log')

plt.xlim(1,1000)
plt.ylim(1e-3,1)
plt.savefig("relicdensity.png")
plt.show()
```
